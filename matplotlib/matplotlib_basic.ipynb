#### Pyplot API & Object-Oriented API
  - Pyplot API - Nhanh chóng
  - Object-Oriented API - Nâng cao hơn

**Pyplot API**

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)
plt.plot(x, y, color="green");
```

```python
# dat tieu de va chu thich cot cho bang
x = np.linspace(0, 10, 1000)
y = np.sin(x)
plt.title("Do Thi Hinh Sin")
plt.xlabel("Bien X")
plt.ylabel("Bien Y")
plt.plot(x, y, color="red");
```

```python
# gioi han truc x va truc y cho bang
x = np.linspace(0, 10, 1000)
y = np.tan(x)
plt.xlim((0, 6))
plt.ylim(-600, 200)
plt.plot(x, y, color="cyan");
```

```python
# thay doi style cua cac duong thang trong do thi 
x = np.linspace(0, 10, 1000)
y = np.cos(x)
plt.plot(x, y, color="orange", linestyle="dashed");
```

```python
# ve nhieu do thi cung mot bang
# plt.legend de show cac label duoc dinh nghia trong plot ra
x = np.linspace(0, 10, 1000)
plt.style.use('dark_background')
plt.title("Sin(X) & Cos(X)")
plt.xlabel("X")
plt.ylabel("F(X)")
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color="red", label="cos(x)");
plt.legend();
```

```python
# cach thay the xlim, ylim trong do thi
x = np.linspace(0, 10, 1000)
y = np.cos(x)
# su dung xlim, ylim
# plt.xlim(2, 8)
# plt.ylim(-1.0, 1.0)
# su dung axis(xmin, xmax, ymin, ymax) cung cho ra ket qua tuong tu
plt.axis([2, 8, -1, 1])
plt.axis("equal")
plt.plot(x, y, color="cyan");
```

**Object-Oriented API**

```python
# Chuan bi du lieu
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Thiet lap plot
fig, ax = plt.subplots(figsize=(5,5)) # figure size = Width x Height of the plot

# Plot tu du lieu
ax.plot(x, y);

ax.set(title="A simple plot",
       xlabel="X-axis",
       ylabel="Y-axis");
```

#### Giới thiệu các loại biểu đồ và đồ thị thông dụng trong Matplotlib

- **line - duong**
- **scatter - phan tan**
- **bar - cot**

**line chart**

```python
# tao mot mang
x = np.linspace(0, 10, 100)
x[:5]
```

**scatter chart**

```python
# ve bieu do phan tan bang Pyplot API
plt.scatter(x, np.exp(x)); # y = e^(x)
```

```python
# ve bieu do phan tan bang OO API

# Chuan bi du lieu
random = np.random.RandomState(0)

x = random.randn(100)
y = random.randn(100)

# dinh nghia kich thuoc
sizes = 1000 * random.randn(100);

# dinh nghia mau sac
colors = random.randn(100);

# Ve bieu do
fig, ax = plt.subplots();

img = ax.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.4);

# Hien thi them thong tin cua tung mau sac
fig.colorbar(img);
```

**bar chart**

  - trục ngang
  - trục dọc

```python
# chuan bi du lieu bang dictionary python
soft_drink_prices = { "Coke": 10,
                      "Pepsi": 12,
                      "Sprite": 9,
                      "C2": 11,
                      "Lavie": 8}

# ve do thi cot bang truc doc OO API
fig, ax = plt.subplots()

ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax.set(title="Bach Hoa Xanh's Sort Drink Prices",
       ylabel="Price (VND)");
```

```python
# ve do thi cot truc ngang OO API
fig, ax = plt.subplots()

ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));

ax.set(title="Bach Hoa Xanh's Sort Drink Prices",
       xlabel="Price (VND)");
```

**subplots**

```python
# TH1: OO API
# chia ra nhieu subplot de ve
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                              ncols=2,
                                              figsize=(15, 10))

# ve tung hinh tren tung subplot tuong ung
# ax1
x1 = np.linspace(0, 10, 100)

ax1.plot(x1, x1+10, color="blue");
ax1.plot(x1, np.cos(x1), color="red");

# ax2
rnd = np.random.RandomState(0)
x2 = random.randn(100)
y2 = random.randn(100)
sizes = 1000 * random.randn(100)
colors = random.randn(100)

ax2.scatter(x2, y2, s=sizes, c=colors);

# ax3
soft_drink_prices = { "Coke": 10,
                      "Pepsi": 12,
                      "Sprite": 9,
                      "C2": 11,
                      "Lavie": 8}

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());

# ax4
np.random.seed(23)
student_height = np.random.normal(165, 10, 500)

ax4.hist(student_height, bins=30);
```

```python
# TH2: OO API
# chia ra nhieu subplot de ve
fig, ax = plt.subplots(nrows=2,
                        ncols=2,
                          figsize=(15, 10))

# ve tung hinh tren tung subplot tuong ung
# ax1
x1 = np.linspace(0, 10, 100)

ax[0,0].plot(x1, x1+10, color="blue");
ax[0,0].plot(x1, np.cos(x1), color="red");

# ax2
rnd = np.random.RandomState(0)
x2 = random.randn(100)
y2 = random.randn(100)
sizes = 1000 * random.randn(100)
colors = random.randn(100)

ax[0,1].scatter(x2, y2, s=sizes, c=colors);

# ax3
soft_drink_prices = { "Coke": 10,
                      "Pepsi": 12,
                      "Sprite": 9,
                      "C2": 11,
                      "Lavie": 8}

ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());

# ax4
np.random.seed(23)
student_height = np.random.normal(165, 10, 500)

ax[1,1].hist(student_height, bins=30);
```

#### Demo matplotlib với dữ liệu thực tế

```python
# goi cac thu vien can thiet
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# doc du lieu tu ban dau 
cities = pd.read_csv(dataset)
```

```python
# xem nhung du lieu ban dau
cities.head()
```

```python
# trich xuat ra du lieu latd = VI DO va longd = KINH DO
latd, longd = cities.latd, cities.longd 
```

```python
# kiem tra du lieu latd = VI DO
latd[:10]
```

```python
# kiem tra du lieu longd = KINH DO
longd[:10]
```

```python
# trich xuat du lieu population = DAN SO va area = DIEN TICH
population, area = cities['population_total'], cities['area_total_km2']
```

```python
# kiem tra du lieu population = DAN SO
population[:10]
```

```python
# kiem tra du lieu area = DIEN TICH
area[:10]
```

```python
# Ve ra bieu do bang Pyplot API
plt.style.use('seaborn')
plt.figure(figsize=(9,7))

plt.scatter(longd, latd, c=np.log10(population), cmap='viridis', s=area, linewidths=0, alpha=0.5);

# danh dau ki hieu ten goi cua tung du lieu
plt.axis('equal');
plt.xlabel('Longatitude');
plt.ylabel('Latitude');
plt.colorbar(label='log$_(10)$(Population)');
plt.clim((3, 7));

# Tao legend cho cac kich co cua cac thanh pho
area_range = [50, 100, 300, 500]

for area in area_range:
  plt.scatter([], [], s=area, c='k', alpha=0.5, label=str(area) + ' km$^2$')

plt.legend(scatterpoints=1, labelspacing=2, title='Area');

plt.title('California Cities: Population & Area Distribution');
```
