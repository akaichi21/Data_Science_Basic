#### II. Push và Pop trong Stack

stack.png

**PUSH** là hành động thêm một phần tử vào trên cùng của ngăn xếp

**POP** là hành động lấy ra một phần tử trên cùng của ngăn xếp

```python
class Stack:
  def __init__(self):
    self.stack = [] # ngăn xếp trống

  # Thêm một phần tử vào trên cùng ngăn xếp
  def push(self, element):
    self.stack.append(element)
    
  # Lấy ra một phần tử trên cùng ngăn xếp
  def pop(self):
    if len(self.stack) > 0:
      self.stack.pop()
    else:
      print("Stack is empty")
    
  # In tất cả các phần tử trong ngăn xếp ra
  def printAllElementStack(self):
    if len(self.stack) > 0:
      for e in self.stack:
        print(e)
    else:
      print("Stack is empty")
```

```python
stack2 = Stack()
```

```python
stack2.stack
```

```python
stack2.printAllElementStack()
```

```python
# thêm phần tử số 21 lên trên cùng ngăn xếp
stack2.push(21)
```

```python
# thêm phần tử số 5 lên trên cùng ngăn xếp
stack2.push(5)
```

```python
stack2.printAllElementStack()
```

```python
# Xóa phần tử trên cùng ngăn xếp, tức phần tử được thêm vào sau cùng sẽ là phần tử trên cùng ngăn xếp
stack2.pop()
```

```python
stack2.printAllElementStack()
```

```python
stack2.pop()
```

```python
stack2.printAllElementStack()
```

#### III. Ứng dụng của Stack
  - Stack được sử dụng nhiều trong các trình biên dịch để biên dịch các ngôn ngữ lập trình
  - Stack được sử dụng để kiểm tra tính chính xác của công thức toán học, công thức hoá học, ví dụ đảm bảo việc đóng mở ngoặc tròn là đầy đủ:
    - Toán học 4 * ( 5 + 10)
    - Hoá học (COO)2H3
  - Stack được sử dụng trong việc xây dựng các máy tính cá nhân đơn giản, Stack giúp quản lý thứ tự các phép tính, phép nào thực hiện trước, phép nào thực hiện sau.
