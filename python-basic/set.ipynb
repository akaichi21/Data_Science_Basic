- Ta có thể dịch Set là tập hợp với các tính chất đặc trưng như :
  - Tập hợp các phần tử này sẽ không có thứ tự
  - Mỗi phần tử phải là duy nhất (Unique) và phần tử phải bất biến (Immutable)
  - Trái ngược lại thì Set là khả biến (Mutable), ta có thể thêm, xóa các phần tử

#### II. Các phương thức trong Set
###### 1. add()
Thêm một phần tử vào Set

```python
set_names = {"Khoa", "Bich", "Hai", "Hoa", "Bao", "Nhu", "Huy", "Oanh"}
```

```python
set_names.add("Thinh")
```

```python
set_names
```

###### 2. remove()
Xóa một phần tử khỏi Set, trả về lỗi nếu phần tử không tồn tại

```python
set_animals = {"Dog", "Cat", "Mouse"}
```

```python
set_animals.remove("Cat")
```

```python
set_animals
```

```python
set_animals.remove("Horse")
```

###### 3. discard()
Xóa một phần tử ra khỏi Set, không trả về lỗi nếu phần tử không tồn tại

```python
set_numbers = {1, 2, 3, 4, 5, 6, 8}
```

```python
set_numbers
```

```python
set_numbers.discard(8)
```

```python
set_numbers
```

```python
set_numbers.discard(7)
```

```python
set_numbers
```

###### 4. union()
Trả về hợp của hai tập hợp

```python
set_a = {1 , 2 , 3 , 4}
```

```python
set_b = {5 , 6 , 7 , 8}
```

```python
set_a.union(set_b)
```

```python
set_b.union(set_a)
```

```python
# tương đương
set_a | set_b
```

###### 5. intesection()
Trả về giao của hai tập hợp

```python
set_a = {3 , 5 , 2 , 6}
```

```python
set_b = {4 , 6 , 1 , 9}
```

```python
set_a.intersection(set_b)
```

```python
set_b.intersection(set_a)
```

```python
# tương đương
set_a & set_b
```

###### 6. difference()
Trả về hiệu hai tập hợp, tức là có ở trước mà không có ở sau

```python
set1 = {"A", "B", "C", "D"}
```

```python
set2 = {"E", "F", "B", "C"}
```

```python
set1.difference(set2)
```

```python
set2.difference(set1)
```

```python
# tương đương
set1 - set2
```

```python
# tương đương
set2 - set1
```

#### III. Frozenset
  - Nếu tuple là một list bất biến thì frozenset là một set bất biến. Ta không thể thêm xóa phần tử trong frozenset

```python
frozenSet = frozenset(["Khoa", "Bich", "Hoa", "Thinh"])
```

```python
frozenSet.add("Bao")
```

```python
frozenSet.remove("Hoa")
```
