#### 2. Enqueue và Dequeue

CqutZ.png

**Enqueue** là thêm một phần tử vào Queue, phần tử này sẽ nằm cuối cùng trong Queue.

**Dequeue** là xóa đi một phần tử của Queue và phần tử này sẽ là phần tử đầu tiên của Queue

```python
class Queue:
  def __init__(self):
    self.queue = []

  # Thêm một phần tử vào Queue
  def enqueue(self, e):
    self.queue.append(e)
    
  # Xóa một phần tử khỏi Queue
  def dequeue(self):
    if len(self.queue) > 0:
      self.queue.pop(0)
    else:
      print("Queue is empty")

  # In ra toàn bộ phần tử trong Queue
  def printAllElement(self):
    if len(self.queue) > 0:
      for e in self.queue:
        print(e)
    else:
      print("Queue is empty")
```

```python
queue2 = Queue()
```

```python
# Thêm 3 phần tử vào Queue
queue2.enqueue("Hoa")
queue2.enqueue("Bich")
queue2.enqueue("Cam")
```

```python
# In tất cả các phần tử trong Queue
queue2.printAllElement()
```

```python
# Xóa phần tử đầu tiên ra khỏi Queue
queue2.dequeue()
```

```python
queue2.printAllElement()
```

#### 3. Sử dụng thư viện

```python
import queue
```

##### Khởi tạo một Queue

```python
q = queue.Queue()
```

##### Thêm một phần tử vào Queue

```python
q.put('Khoa')
q.put('Bao')
q.put('Hai')
q.put('Huy')
```

##### Kiểm tra một phần tử trong Queue

```python
q.qsize()
```

##### Xóa một phần tử khỏi Queue

```python
q.get()
```

##### Lấy giá trị đầu tiên của Queue

```python
q.queue[0]
```
